##### 🚨 The Problem

---

###### 💥 Why This Happens

You defined:

```python
@Amount.setter
def Credit(self, AmtToCredit):
```

But in Python:

👉 A `@property` setter **does NOT create a callable method**
👉 It creates an **assignment-style interface**, not a function

So instead of:

```python
r0908779.Credit(500000) ❌
```

Python expects:

```python
r0908779.Amount = value  ✅
```

---

###### ⚠️ Bigger Issue in Your Code

You actually did something invalid:

```python
@Amount.setter
def Debit(...)

@Amount.setter
def Credit(...)
```

👉 You **cannot have two setters for the same property (`Amount`)**

Only **one setter per property** is allowed.

---

##### ✅ Correct Approaches

✔️ Option 1: Use Normal Methods (BEST for Banking Logic)

This is the cleanest and most realistic design.

```python
class BankAccount:
    def __init__(self, HolderName, Age, Sex, Amount, Currency):
        self.__HolderName = HolderName
        self.__Age = Age
        self.__Sex = Sex
        self.__Amount = Amount
        self.Currency = Currency

    @property
    def Amount(self):
        return f"{self.Currency}{self.__Amount}"

    def debit(self, amt):
        if self.__Amount >= amt:
            self.__Amount -= amt
        else:
            print("Insufficient balance")

    def credit(self, amt):
        self.__Amount += amt
```

Usage:

```python
acc = BankAccount('Max', 21, 'M', 50000000, '$')

acc.credit(500000)
acc.debit(200000)

print(acc.Amount)
```

---

✔️ Option 2: Use Setter Properly (Less Practical Here)

If you *really* want setter:

```python
@Amount.setter
def Amount(self, value):
    self.__Amount = value
```

Usage:

```python
acc.Amount = 100000
```

But❗ this **does NOT distinguish debit vs credit**, so it's not ideal.

---

###### Key Takeaways (Important for Exams)

* `@property` → accessed like a variable
* `@property.setter` → used with assignment, NOT function call
* Only **one setter per property**
* For operations like debit/credit → use **methods, not setters**

````python
#Getter() and normal methods for access

class BankAccountR:
    def __init__(self, HolderName, Age, Sex, Amount, Currency):
        self.__HolderName=HolderName
        self.__Age=Age
        self.__Sex=Sex
        self.__Amount=Amount
        self.Currency=Currency

    @property
    def HolderName(self):
        return self.__HolderName
    
    @property
    def Age(self):
        return self.__Age
    
    @property
    def Sex(self):
        return self.__Sex
    
    @property
    def Amount(self):
        return f"{self.Currency}{self.__Amount}"
    

    def debit(self, amt):
        if self.__Amount >= amt:
            self.__Amount -= amt
        else:
            print("Insufficient balance")

    def credit(self, amt):
        self.__Amount += amt
        print(f"{self.__HolderName}'s account credited with {amt}, new balance {self.__Amount}")
````
