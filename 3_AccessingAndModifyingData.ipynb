```python
class Company:
    def __init__(self, name):
        self.name=name

class Employee:
    def __init__(self, name, email, ID, Company):
        self.__name=name
        self.__email=email
        self.__eID=ID
        self.__Company = Company
        
    
    def constructEmail(self):
        self.email = self.name.lower() +'@' + self.Company.name.lower() + '.com'
        print(self.email)


    # def getName(self):
    #     return self.__name
    
    # def getEmail(self):
    #     return self.__email
    
    # def getID(self):
    #     return self.__eID
    
    # def getCompany(self):
    #     return self.__Company.name


    @property
    def email(self):
        print("Email accessed. ")
        return self.__email
    

    @email.setter
    def email(self, newEmail):
        if '@' and '.' in newEmail:
            self.__email=newEmail
            print(f"Email updated to: {self.__email}")
```
