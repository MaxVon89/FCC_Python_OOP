```python
class Person:
    def __init__(self, name, mother, father, brother, sister, spouse):
        self.name=name
        self.mother=mother
        self.father=father
        self.brother=brother
        self.sister=sister
        self.spouse=spouse

    def printFamilyTree(self):
        print(f"""{self.father.name} <-> {self.mother.name}""")
        if self.brother:
            print(f"""|             |
                  {self.name}""")
            for i in self.brother:
                print(f"                {self.brother[i].name}")
        
        elif self.sister:
            print(f"""|             |
{self.name}""")

            for i in self.sister:
                print(f"                {self.sister[i].name}")            

        elif self.brother and self.sister:
            print(f"""|             |               |
                  {self.name}   {self.sister.name}    {self.brother.name}
""")
        else:
            print(f"""|
                  {self.name}
""")
```

```python
Mike = Person('Mike', False, False, False, False, False)
Jane = Person('Jane', False, False, False, False, False)
Max = Person('Max', False, False, False, False, False)
Ashley = Person('Ashley', False, False, False, False, False)
Morgan = Person('Morgan', False, False, False, False, False)

Mike.spouse = Jane
Jane.spouse = Mike

Max.mother = Jane
Max.father = Mike

Ashley.mother = Jane
Ashley.father = Mike

Morgan.mother = Jane
Morgan.father = Mike

Max.sister = {1:Ashley, 2:Morgan}

Ashley.brother = Max
Morgan.brother = Max

Max.printFamilyTree()
```
