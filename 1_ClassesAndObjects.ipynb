```python
name='Something'
age=21

print(type(name))
print(type(age))
```

These vars. name and age are objects of 'str' and 'int' classes resp., and the type is decided by the Python interpreter.

```python
#Creating a class

class Dog:
    def bark(self):
        print("BARK!")


cheems = Dog()

cheems.bark
#Output:
# >> <bound method Dog.bark of <__main__.Dog object at 0x00000243CB62EE40>>

cheems.bark()
#Output
# >> BARK!
```

```python
class Person:
    def __init__(self, Name):
        self.Name=Name
```

```python
Me = Person('Max')
```

```python
class Vehicle:
    def __init__(self, Type, Manufc, Model, Ownership, Owner):
        self.Type=Type
        self.Manufc=Manufc
        self.Model=Model
        self.Ownership=Ownership
        self.Owner=Owner

    def printSpecs(self):
        print(self.Type)
        print(self.Manufc)
        print(self.Model)
        print(self.Ownership)
        # print(self.Owner.Name()) #Wrong
        print(self.Owner.Name)
```

```python
myTruck = Vehicle('Truck', 'Chevy', 'F150', True, Me)
myTruck.printSpecs()
```
